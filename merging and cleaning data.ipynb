```python
import pandas as pd
import numpy as np
```

```python
raters=pd.read_csv('raters.csv', index_col=[0])
```

```python
raters.head()
```

```python
raters.shape
```

```python
def split(string): #define a function to split "Beer Id" to merge beers and raters together base on beer profile
    idbeer=string.rsplit("?ba")
    return idbeer[0]
```

```python
split('/beer/profile/182/1351/?ba=BarfyMan362#review')
```

```python
raters['Url']=raters['Beer Id'].map(lambda x : split(x))
```

```python
raters.head()
```

```python
raters.shape
```

```python
beers=pd.read_csv('beers.csv', index_col=[0])
```

```python
beers
```

```python
beers.head()
```

```python
#merge beers and raters data base on raters
data=pd.merge(beers,raters, on='Url', how='left')
```

```python
data.head()
```

```python
#checking null values
data.isnull().sum()
```

```python
#drop Null values
data.dropna(axis=0, inplace=True)
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
#checking for dupclicates
data.duplicated().sum()
```

```python
data
```

```python
#set my data index after drop dupclicates
data.index=np.arange(1, len(data)+1)
```

```python
data.drop('Beer Id', axis=1, inplace=True)
```
